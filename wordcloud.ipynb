### **Importing the libraries**

```python
import wikipedia
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
```

### **For Collecting the page contents from wikipedia based on the keyword is given**

```python
def wikipediaFunc():    
    x=input("Enter the title(keyword): ")
    title=wikipedia.search(x)[0]
    page=wikipedia.page(title)
    text=page.content
    return text
```

### **Visualizing the images**

```python
def visualizing_images(source_img,generated_img):
    plt.figure(figsize=(15,5))
    plt.subplot(1,2,1)
    plt.title('Source Image')
    img1=plt.imread(source_img)
    plt.imshow(img1)

    plt.subplot(1,2,2)
    plt.title('Generated Image')
    img2=plt.imread(generated_img)
    plt.imshow(img2)

    plt.show()
```

### **Main Function that generates the wordcloud image**

```python
def Word_Cloud_Generator(source_img,destination_path):
    
    text=wikipediaFunc()
    stopwordset=set(STOPWORDS)
    #stopwordset.remove("-")

    background=np.array(Image.open(source_img))
    wc=WordCloud(background_color="black", max_words=1000, mask=background,
                   stopwords=stopwordset, contour_width=4, contour_color='white')       #     background_color='white',mask=background,stopwords=stopwordset,inter)

    wc.generate(text)
    wc.to_file(destination_path+"/generated.png")
    
    ######################
    visualizing_images(source_img,destination_path+"/generated.png")
```

```python
Word_Cloud_Generator(r"cloud.png",r".")
```

```python
Word_Cloud_Generator(r"Capture1.png",r".")
```
